```python
import random
from math import inf
import threading
import time

player, opponent = 'x', 'o'
value = 0
nodes_visited_sequential = 0
nodes_visited_parallel = 0

def isMovesLeft(board):
    for i in range(3):
        for j in range(3):
            if board[i][j] == '_':
                return True
    return False

def evaluate(b):
    if not isMovesLeft(b):
        return 0

    for row in range(3):
        if b[row][0] == b[row][1] and b[row][1] == b[row][2]:
            if b[row][0] == player:
                return 10
            elif b[row][0] == opponent:
                return -10

    for col in range(3):
        if b[0][col] == b[1][col] and b[1][col] == b[2][col]:
            if b[0][col] == player:
                return 10
            elif b[0][col] == opponent:
                return -10

    if b[0][0] == b[1][1] and b[1][1] == b[2][2]:
        if b[0][0] == player:
            return 10
        elif b[0][0] == opponent:
            return -10

    if b[0][2] == b[1][1] and b[1][1] == b[2][0]:
        if b[0][2] == player:
            return 10
        elif b[0][2] == opponent:
            return -10

    return 0

def minimax_alpha_beta(board, depth, alpha, beta, isMax, start_idx, end_idx):
    global nodes_visited_sequential, nodes_visited_parallel
    nodes_visited_sequential += 1
    if isMax:
        nodes_visited_parallel += 1

    score = evaluate(board)
    if score == 10:
        return score
    if score == -10:
        return score
    if not isMovesLeft(board):
        return 0

    if isMax:
        best = -1000
        for i in range(3):
            for j in range(3):
                if board[i][j] == '_' and (i * 3 + j) >= start_idx and (i * 3 + j) < end_idx:
                    board[i][j] = player
                    val = minimax_alpha_beta(board, depth + 1, alpha, beta, False, start_idx, end_idx)
                    best = max(best, val)
                    alpha = max(alpha, best)
                    board[i][j] = '_'
                    if beta <= alpha:
                        return best
        return best
    else:
        best = 1000
        for i in range(3):
            for j in range(3):
                if board[i][j] == '_' and (i * 3 + j) >= start_idx and (i * 3 + j) < end_idx:
                    board[i][j] = opponent
                    val = minimax_alpha_beta(board, depth + 1, alpha, beta, True, start_idx, end_idx)
                    best = min(best, val)
                    beta = min(beta, best)
                    board[i][j] = '_'
                    if beta <= alpha:
                        return best
        return best

def findBestMove(board):
    global nodes_visited_sequential, nodes_visited_parallel
    bestVal = -1000
    bestMove = (-1, -1)

    # Sequential version
    start_time = time.time()
    start_idx = 0
    end_idx = 9
    moveVal = minimax_alpha_beta(board, 0, -1000, 1000, False, start_idx, end_idx)
    if moveVal > bestVal:
        bestVal = moveVal

    sequential_time = time.time() - start_time

    # Parallel version
    bestVal = -1000
    bestMove = (-1, -1)

    results = [None, None]

    def worker(start_idx, end_idx):
        best_val = -1000
        best_move = (-1, -1)
        for i in range(3):
            for j in range(3):
                if board[i][j] == '_' and (i * 3 + j) >= start_idx and (i * 3 + j) < end_idx:
                    board[i][j] = player
                    moveVal = minimax_alpha_beta(board, 0, -1000, 1000, False, start_idx, end_idx)
                    board[i][j] = '_'
                    if moveVal > best_val:
                        best_move = (i, j)
                        best_val = moveVal
        results[start_idx // 4] = (best_val, best_move)

    start_time = time.time()
    thread1 = threading.Thread(target=worker, args=(0, 5))
    thread2 = threading.Thread(target=worker, args=(5, 9))
    thread1.start()
    thread2.start()
    thread1.join()
    thread2.join()
    parallel_time = time.time() - start_time

    for result in results:
        if result[0] > bestVal:
            bestVal = result[0]
            bestMove = result[1]

    print(f"Sequential version value: {bestVal}")
    print(f"Sequential version time: {sequential_time:.6f} seconds")
    print(f"Sequential version nodes visited: {nodes_visited_sequential}")
    print("The Optimal Move is :")
    print("ROW:", bestMove[0], " COL:", bestMove[1])
    print()
    print("=============================================")

    print(f"Parallel version value: {bestVal}")
    print(f"Parallel version time: {parallel_time:.6f} seconds")
    print(f"Parallel version nodes visited: {nodes_visited_parallel}")
    print("The Optimal Move is :")
    print("ROW:", bestMove[0], " COL:", bestMove[1])
    print()

    return bestMove

def generate_random_board():
    symbols = ['x', 'o', '_']
    while True:
        board = [[random.choice(symbols) for _ in range(3)] for _ in range(3)]
        return board

for i in range(3):
    board = generate_random_board()

    print("RUN : ",i)
    print("Random starting position:")
    for row in board:
        print(row)
    print()

    findBestMove(board)
    print("---------------------------------------------------")
```

```python
import random
from math import inf

player, opponent = 'x', 'o'

def isMovesLeft(board):
    for i in range(3):
        for j in range(3):
            if board[i][j] == '_':
                return True
    return False

def evaluate(b):
    if not isMovesLeft(b):
        return 0

    for row in range(3):
        if b[row][0] == b[row][1] and b[row][1] == b[row][2]:
            if b[row][0] == player:
                return 10
            elif b[row][0] == opponent:
                return -10

    for col in range(3):
        if b[0][col] == b[1][col] and b[1][col] == b[2][col]:
            if b[0][col] == player:
                return 10
            elif b[0][col] == opponent:
                return -10

    if b[0][0] == b[1][1] and b[1][1] == b[2][2]:
        if b[0][0] == player:
            return 10
        elif b[0][0] == opponent:
            return -10

    if b[0][2] == b[1][1] and b[1][1] == b[2][0]:
        if b[0][2] == player:
            return 10
        elif b[0][2] == opponent:
            return -10

    return 0

def heuristic_evaluation(board, player, opponent):
    player_winning_moves = 0
    opponent_winning_moves = 0

    # Check rows
    for row in range(3):
        if board[row].count(player) == 2 and '_' in board[row]:
            player_winning_moves += 1
        elif board[row].count(opponent) == 2 and '_' in board[row]:
            opponent_winning_moves += 1

    # Check columns
    for col in range(3):
        column = [board[row][col] for row in range(3)]
        if column.count(player) == 2 and '_' in column:
            player_winning_moves += 1
        elif column.count(opponent) == 2 and '_' in column:
            opponent_winning_moves += 1

    # Check diagonals
    diagonal1 = [board[i][i] for i in range(3)]
    diagonal2 = [board[i][2 - i] for i in range(3)]
    if diagonal1.count(player) == 2 and '_' in diagonal1:
        player_winning_moves += 1
    elif diagonal1.count(opponent) == 2 and '_' in diagonal1:
        opponent_winning_moves += 1
    if diagonal2.count(player) == 2 and '_' in diagonal2:
        player_winning_moves += 1
    elif diagonal2.count(opponent) == 2 and '_' in diagonal2:
        opponent_winning_moves += 1

    return player_winning_moves - opponent_winning_moves

def minimax(board, depth, isMax):
    score = evaluate(board)
    if score == 10:
        return score, 1
    if score == -10:
        return score, 1
    if not isMovesLeft(board):
        return 0, 1

    nodes_visited = 0
    if isMax:
        best = -1000
        for i in range(3):
            for j in range(3):
                if board[i][j] == '_':
                    board[i][j] = player
                    val, visited = minimax(board, depth + 1, not isMax)
                    nodes_visited += visited
                    best = max(best, val)
                    board[i][j] = '_'
        return best, nodes_visited
    else:
        best = 1000
        for i in range(3):
            for j in range(3):
                if board[i][j] == '_':
                    board[i][j] = opponent
                    val, visited = minimax(board, depth + 1, not isMax)
                    nodes_visited += visited
                    best = min(best, val)
                    board[i][j] = '_'
        return best, nodes_visited

def minimax_alpha_beta(board, depth, alpha, beta, isMax):
    score = evaluate(board)
    if score == 10:
        return score, 1
    if score == -10:
        return score, 1
    if not isMovesLeft(board):
        return 0, 1

    nodes_visited = 0
    if isMax:
        best = -1000
        for i in range(3):
            for j in range(3):
                if board[i][j] == '_':
                    board[i][j] = player
                    val, visited = minimax_alpha_beta(board, depth + 1, alpha, beta, False)
                    nodes_visited += visited
                    best = max(best, val)
                    alpha = max(alpha, best)
                    board[i][j] = '_'
                    if beta <= alpha:
                        return best, nodes_visited
        return best, nodes_visited
    else:
        best = 1000
        for i in range(3):
            for j in range(3):
                if board[i][j] == '_':
                    board[i][j] = opponent
                    val, visited = minimax_alpha_beta(board, depth + 1, alpha, beta, True)
                    nodes_visited += visited
                    best = min(best, val)
                    beta = min(beta, best)
                    board[i][j] = '_'
                    if beta <= alpha:
                        return best, nodes_visited
        return best, nodes_visited

def minimax_alpha_beta_heuristic(board, depth, alpha, beta, isMax, player, opponent):
    score = evaluate(board)
    if score == 10:
        return score, 1
    if score == -10:
        return score, 1
    if not isMovesLeft(board):
        return 0, 1

    nodes_visited = 0
    if isMax:
        best = -1000
        for i in range(3):
            for j in range(3):
                if board[i][j] == '_':
                    board[i][j] = player
                    heuristic_value = heuristic_evaluation(board, player, opponent)
                    val, visited = minimax_alpha_beta_heuristic(board, depth + 1, alpha, beta, False, player, opponent)
                    nodes_visited += visited
                    val += heuristic_value
                    best = max(best, val)
                    alpha = max(alpha, best)
                    board[i][j] = '_'
                    if beta <= alpha:
                        return best, nodes_visited
        return best, nodes_visited
    else:
        best = 1000
        for i in range(3):
            for j in range(3):
                if board[i][j] == '_':
                    board[i][j] = opponent
                    heuristic_value = heuristic_evaluation(board, player, opponent)
                    val, visited = minimax_alpha_beta_heuristic(board, depth + 1, alpha, beta, True, player, opponent)
                    nodes_visited += visited
                    val -= heuristic_value
                    best = min(best, val)
                    beta = min(beta, best)
                    board[i][j] = '_'
                    if beta <= alpha:
                        return best, nodes_visited
        return best, nodes_visited

def findBestMove(board):
    bestVal = -1000
    bestMove = (-1, -1)
    nodes_visited = 0

    # Standard minimax
    for i in range(3):
        for j in range(3):
            if board[i][j] == '_':
                board[i][j] = player
                moveVal, visited = minimax(board, 0, False)
                nodes_visited += visited
                board[i][j] = '_'
                if moveVal > bestVal:
                    bestMove = (i, j)
                    bestVal = moveVal
    print("Minimax:")
    print("The value of the best Move is :", bestVal)
    print("Size Of Search Space:", nodes_visited)
    print("The Optimal Move is :")
    print("ROW:", bestMove[0], " COL:", bestMove[1])
    print()
    print("==========================================")

    # Standard alpha-beta pruning
    bestVal = -1000
    bestMove = (-1, -1)
    nodes_visited = 0
    for i in range(3):
        for j in range(3):
            if board[i][j] == '_':
                board[i][j] = player
                moveVal, visited = minimax_alpha_beta(board, 0, -1000, 1000, False)
                nodes_visited += visited
                board[i][j] = '_'
                if moveVal > bestVal:
                    bestMove = (i, j)
                    bestVal = moveVal
    print("Alpha-Beta Pruning:")
    print("The value of the best Move is :", bestVal)
    print("Size Of Search Space:", nodes_visited)
    print("The Optimal Move is :")
    print("ROW:", bestMove[0], " COL:", bestMove[1])
    print()

    print("==========================================")

    # Heuristic-based alpha-beta pruning
    bestVal = -1000
    bestMove = (-1, -1)
    nodes_visited = 0
    for i in range(3):
        for j in range(3):
            if board[i][j] == '_':
                board[i][j] = player
                moveVal, visited = minimax_alpha_beta_heuristic(board, 0, -1000, 1000, False, player, opponent)
                nodes_visited += visited
                board[i][j] = '_'
                if moveVal > bestVal:
                    bestMove = (i, j)
                    bestVal = moveVal

                    if(bestVal<0):
                      bestVal=-10

    print("Heuristic-based Alpha-Beta Pruning:")
    print("The value of the best Move is :", bestVal)
    print("Size Of Search Space:", nodes_visited)
    print("The Optimal Move is :")
    print("ROW:", bestMove[0], " COL:", bestMove[1])
    print()
    print("==========================================")
    print("---------------------------------------------------")
    print()

    return bestMove

def generate_random_board():
    symbols = ['x', 'o', '_']
    while True:
        board = [[random.choice(symbols) for _ in range(3)] for _ in range(3)]
        return board

for i in range(3):
    # Driver code
    board = generate_random_board()
    print("Random starting position:")
    for row in board:
        print(row)
    print()

    findBestMove(board)
```

# ->  4) Develop a tic-tac-toe game, where the AI is a “weak” player. Modify the game in the following manner:

```python
from random import choice

def print_board(board):
    print(f"{board[0]}|{board[1]}|{board[2]}")
    print("-"*5)
    print(f"{board[3]}|{board[4]}|{board[5]}")
    print("-"*5)
    print(f"{board[6]}|{board[7]}|{board[8]}")

def check_win(board, mark):
    return (
        (board[0] == board[1] == board[2] == mark) or
        (board[3] == board[4] == board[5] == mark) or
        (board[6] == board[7] == board[8] == mark) or
        (board[0] == board[3] == board[6] == mark) or
        (board[1] == board[4] == board[7] == mark) or
        (board[2] == board[5] == board[8] == mark) or
        (board[0] == board[4] == board[8] == mark) or
        (board[2] == board[4] == board[6] == mark)
    )

def o_comp(board, avoid_tile):
    available_moves = [i for i in range(9)]
    if avoid_tile in available_moves:
        available_moves.remove(avoid_tile)

    if len(available_moves) == 0:
        return

    move = choice(available_moves)
    board[move] = 'O'

def pvc():
    print("="*50)
    print("TIC-TAC-TOE with a Weak AI Player")
    print("="*50)
    print("Enter to play 1st or 2nd: ")
    player = int(input())
    board = [' ' for _ in range(9)]

    if player == 1:
        avoid_tile = choice(range(9))
        print(f"Avoiding tile: {avoid_tile}")
        print("Player goes first (X)")
    else:
        avoid_tile = -1
        print("Computer goes first (O)")

    while ' ' in board:
        print_board(board)

        if player == 1:
            move = int(input("Enter a number between 1-9: ")) - 1
            if board[move] != ' ':
                print("Invalid move! Try again.")
                continue
            board[move] = 'X'
        else:
            o_comp(board, avoid_tile)

        if check_win(board, 'X'):
            print_board(board)
            print("Player wins!")
            break
        elif check_win(board, 'O'):
            print_board(board)
            print("Computer wins!")
            break
        elif ' ' not in board:
            print_board(board)
            print("It's a tie!")
            break

        if player == 2:
            move = int(input("Enter a number between 1-9: ")) - 1
            if board[move] != ' ':
                print("Invalid move! Try again.")
                continue
            board[move] = 'X'
        else:
            o_comp(board, avoid_tile)

        if check_win(board, 'X'):
            print_board(board)
            print("Player wins!")
            break
        elif check_win(board, 'O'):
            print_board(board)
            print("Computer wins!")
            break
        elif ' ' not in board:
            print_board(board)
            print("It's a tie!")
            break

if __name__ == "__main__":
    pvc()
```
